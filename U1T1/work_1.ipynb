# 2.1 Calculating Coin Toss Probabilities

We will calculate the probabilities of all possible outcomes of tossing two loaded coins, based on the given probabilities.

## Steps to Solve the Exercise

1. **Define the Probabilities**: We will start by defining the probabilities of getting heads (H) for each of the coins.
2. **Calculate the Results**: We will use the probability formulas to calculate each of the possible results:
- H-H
- H-T
- T-H
- T-T
3. **Organize the Results**: Finally, we will organize the results in a table to make them easier to visualize.

```python
# Step 1: Define the Probabilities
probabilities = [
    (0.5, 0.5),
    (0.6, 0.7),
    (0.4, 0.8),
    (0.1, 0.2),
    (0.3, 0.4)
]
```

```python
# Step 2: Calculate the Results
results = []
for p1, p2 in probabilities:
    H_H = p1 * p2
    H_T = p1 * (1 - p2)
    T_H = (1 - p1) * p2
    T_T = (1 - p1) * (1 - p2)
    results.append((H_H, H_T, T_H, T_T))
```

```python
# Step 3: Organize the Results
results
```

# 2.2 Bayes' Theorem for an Inbox Spam Filter

We will utilize the Baye's Theorem to calculate how many emails in the spam box are not suspected to be spam in flagged emails.

## Steps to Solve the Exercise

1. **Calculate P(F)**: We will calculate the probability of be Flagged
2. **Calculate P(N∣F)P(N∣F)**: 
3. **Calculate the expected number of non-spam emails in the spam box**: Finally, we will organize the results in a table to make them easier to visualize.

**Known probabilities**:
  - \( P(S) = 0.6 \) (60% of emails are spam)
  - \( P(N) = 0.4 \) (40% of emails are not spam)
  - \( P(F | S) = 0.98 \) (98% chance of correctly flagging spam)
  - \( P(F | N) = 0.02 \) (2% chance of incorrectly flagging non-spam as spam)

```python
# Step 1: Calculate P(F)

# Known values
P_S = 0.6  # P(S)
P_N = 0.4  # P(N)
P_F_given_S = 0.98  # P(F | S)
P_F_given_N = 0.02  # P(F | N)
total_spam_emails = 963
```

```python
# Step 2: Calculate P(N | F)
P_N_given_F = (P_F_given_N * P_N) / P_S
P_N_given_F
```

```python
# Step 3: Calculate the expected number of non-spam emails in the spam box

expected_non_spam = total_spam_emails * P_N_given_F
expected_non_spam
```

# 2.3 Markov process string
We will reconstruct the correct answer of the string.

## Steps to Solve the Exercise

1. **Identify the Shift**: A backward shift of 21 is equivalent to a forward shift of 26−21=526−21=5.
2. **Decoding the String**: For each letter in "OCZ XJMMZXO VINRZM", shift forward by 5.

**Known informations**:
  - This is a Caesar cipher with shift 7 applied three times
  - Caesar cipher is a Markov process
  - [Rules](https://www.networkatlas.eu/exercises/2/3/data.txt)


```python
# Step 1 : Apply the known shift on a function using Caesar cipher

encoded_string = "OCZ XJMMZXO VINRZM"
# Shift amount for decoding (forward by 5)
shift_amount = 5

def caesar_decode(input_string: str, shift: int) -> str:
    decoded_string = ""
    
    for char in input_string:
        if char.isalpha():  # Check if the character is a letter
            # Find the base ASCII code for uppercase letters
            base = ord('A')
            # Calculate the original position, apply the shift, and wrap around using modulo
            new_char = chr((ord(char) - base + shift) % 26 + base)
            decoded_string += new_char
        else:
            # Keep spaces as is
            decoded_string += char
            
    return decoded_string
```

```python
# Step 2: Decoding string

# Decode the string
decoded_string = caesar_decode(encoded_string, shift_amount)
decoded_string
```

# Conclusion
With this, we conclude that <strong>THE CORRECT ANSWER</strong> is the decoded string.

# 2.5 Independently Encoded Bits

## Steps to Solve the Exercise

1. **Import the libraries**: Import pandas, scipy and sklearn.
2. **Load the data**: Load the data from exercise.
3. **Calculate**

```python
# Step 1: Import the libraries

import pandas as pd
from scipy.stats import entropy
from sklearn.metrics import mutual_info_score
```

```python
# Step 2: Load the data

df = pd.read_csv("assets/data.txt", sep = "\t")
```

```python
# Step 3: Calculate
v1_ps = (df.groupby(by = "v1").size() / df.shape[0]).values
v2_ps = (df.groupby(by = "v2").size() / df.shape[0]).values

v1_info = entropy(v1_ps)
v2_info = entropy(v2_ps)

print(v1_info)
print(v2_info)

mi = mutual_info_score(df["v1"], df["v2"])

print("Knowing v2 gives us %1.2f%% of v1" % (100 * mi / v1_info))
```

# 4.2 Determine the type of multi-layer network

## Steps to Solve the Exercise

1. **Import libraries**
2. **Read the data**
3. **Determine the layers quantity**
4. **Get the type of the network**

**Known information:**
- in the Chain coupling, all nodes can have at most two coupling edges
- in the Clique coupling, all nodes must have more than one coupling edge
- in the star coupling, all nodes must have either one edge, or |L| - 1 coupling edges

```python
# Step 1: Import libraries

import networkx as nx
from collections import defaultdict
```

```python
# Step 2: Load the data as MultiGraph
G = nx.read_edgelist("assets/multilayer_network_data.txt", create_using = nx.MultiGraph(), data = [("edgetype", str),])
```

```python
# Step 3: Determine how many layers we have
layers = len(set(e[2]["edgetype"] for e in G.edges(data = True))) - 1
```

```python
# Step 4: Get the type of netowkr based on known information
node_cedges = defaultdict(int)
for e in G.edges(data = True):
   if e[2]["edgetype"] == "C":
      node_cedges[e[0]] += 1
      node_cedges[e[1]] += 1

if max(node_cedges.values()) == 2:
   print("Chain coupling")

if min(node_cedges.values()) > 1:
   print("Clique coupling")

if set([1, layers - 1]) == set(node_cedges.values()):
   print("Star coupling")
```

# 4.3 Transform the Hypergraph network

## Steps to Solve the Exercise

1. **Import libraries**
2. **Read the data**
3. **Make the unipartite graph**
4. **Make the bipartite graph**
5. **Show the graphs**

```python
# Step 1: Load libraries

import networkx as nx
from itertools import combinations
```

```python
# Step 2: Load the data.
hyperedges = []
with open("assets/hypergraph_network_data.txt", 'r') as f:
   for line in f:
      hyperedges.append(line.strip().split())
```

```python
# Step 3: Make it into a unipartite graph
G_uni = nx.Graph()
for hyperedge in hyperedges:
   for combination in combinations(hyperedge, 2):
      G_uni.add_edge(*combination)
```

```python
# Step 4: Make it into a bipartite graph
G_bip = nx.Graph()
hyperid = 1
for hyperedge in hyperedges:
   for node in hyperedge:
      G_bip.add_edge(node, "h%s" % hyperid)
   hyperid += 1
```

```python
# Step 5: Show the graphs

nx.draw(G_uni, with_labels = True)
plt.savefig("assets/unipartite_hypergraph_network.png")
plt.show()
```

```python
nx.draw(G_bip, with_labels = True)
plt.savefig("assets/unipartite_hypergraph_network.png")
plt.show()
```

# 4.4 Aggregate the network edges

## Steps to Solve the Exercise

1. **Import libraries**
2. **Read the data**
3. **Group the windows**
4. **Show the graphs**

```python
# Step 1: Load libraries

import networkx as nx
```

```python
# Step 2: Load the data as MultiGraph
G = nx.read_edgelist("assets/dynamic_network_data.txt", create_using = nx.MultiGraph(), data = [("start", int), ("end", int)])
```

```python
# Step 3: Group the windows as 1-3, 4-6, and 7-9 snapshots.
first_window = nx.Graph()
second_window = nx.Graph()
third_window = nx.Graph()
for e in G.edges(data = True):
   if e[2]["start"] <= 3:
      first_window.add_edge(e[0], e[1])
   if e[2]["start"] <= 6 and e[2]["end"] > 3:
      second_window.add_edge(e[0], e[1])
   if e[2]["end"] > 6:
      third_window.add_edge(e[0], e[1])
```

```python
# Step 4: Draw windows

nx.draw(first_window, with_labels = True)
plt.savefig("assets/first_window.png")
plt.show()
```

```python
nx.draw(second_window, with_labels = True)
plt.savefig("assets/second_window.png")
plt.show()
```

```python
nx.draw(third_window, with_labels = True)
plt.savefig("assets/third_window.png")
plt.show()
```
